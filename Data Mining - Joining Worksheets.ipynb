# Importing main Librarys

```python
import pandas as pd

from tkinter import Tk     # from tkinter import Tk for Python 3.x
from tkinter.filedialog import askopenfilename
```

```python
path = askopenfilename() # show an "Open" dialog box and return the path to the selected file
```

# Separating the directory from the file

```python
filename = []
listaPath = list(path)

while True:
    letra = listaPath.pop()
    if letra != '/':
        filename.append(letra)
    else:
        break 
        
filename.reverse()
filename = ''.join(filename)
listaPath = ''.join(listaPath)
```

```python
import os 
os.chdir(listaPath)
arquivos_em_diretorio = os.listdir(listaPath)
```

# Showing files in directory

```python
vetor_dfs = []


for filename in arquivos_em_diretorio:
    flag = 0
    print(filename)
    #arquivo = dt.fread(filename, encoding='latin-1',header=None).to_pandas()
    arquivo = pd.read_csv(filename, encoding='latin-1',header=None, skiprows=1) # Original file is in Portuguese
    
    tam = arquivo.shape[0]
    print(f'Tamnho do DataSet = {tam}')
    
    
    lista_atributos = arquivo.columns.values
    for valor in lista_atributos:
        dados_notNULL = arquivo[valor].notnull().sum()
        if dados_notNULL < tam:
            print('\n\t!existem itens faltantes na planilha!\n')
            flag = 1
    if flag == 0:
        print('OK')
    
    vetor_dfs.append(arquivo)

X_all = pd.concat([vetor_dfs[0]], axis = 0)
display(X_all)
for i in range(1,len(arquivos_em_diretorio)):
    X_all = pd.concat([X_all, vetor_dfs[i]], axis = 0)
    display(X_all)
   
```

```python
X_all.to_csv(listaPath + '/CompleteTable.csv', encoding='Latin-1', header=None, index=False)
```
